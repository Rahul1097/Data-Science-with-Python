```python
#merge operations in pandas
dfExample1 = pd.DataFrame([[4,'QA'],[1,'HR'],[3,'Dev'],[2,'Ops']] ,
columns=['eid','dept'])
dfExample2 = pd.DataFrame([[1,'Prashant'],[2,'Gokul'],[3,'Guna']] ,
columns=['eid','ename'])
```

```python
dfExample1
```

```python
dfExample2
```

```python
#merge #inner join-common elements between both tables
resultDF = pd.merge(dfExample1,dfExample2,on='eid',how='inner')
resultDF
```

```python
#left join
resultDF = pd.merge(dfExample1,dfExample2,on='eid',how='left')
resultDF
```

```python
dept = pd.DataFrame([[4,'QA'],[1,'HR'],[3,'Dev'],[2,'Ops']] , columns=['eid','dept'])
emp = pd.DataFrame([[1,'Prashant'],[2,'Gokul'],[3,'Guna']] , columns=['empid','ename'])
```

```python
dept
```

```python
emp
```

```python
#merge with different column names
resultDF2 = pd.merge(emp,dept,left_on='empid',right_on ='eid')
resultDF2
```

```python
#take the index of the record-in this case last record
resultDF2.tail(1).index.item()
```

```python
#Concatenation
empExample = pd.DataFrame([[1,'Prashant'],[2,'Gokul'],[3,'Guna']] , 
columns=['empid','ename'])
empExample2 = pd.DataFrame([[4,'Nik'],[5,'Ashish'],[6,'Asha']] , 
columns=['empid','ename'])
empExample 
```

```python
empExample2
```

```python
#Row-wise concatenation 
resultEmp=pd.concat([empExample,empExample2],axis=0)#column indexes are same in DFs
resultEmp
```

```python
resultEmp.iloc[[0,2],:]
```

```python
resultEmp.reset_index(inplace = True,drop = True)
```
