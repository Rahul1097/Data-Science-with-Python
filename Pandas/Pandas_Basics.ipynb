```python
#add yearly salary column
yearlysalary = employeedf_from_file.empSalary *1.2
yearlysalary
```

```python
#updating existing column
employeedf_from_file['New Salary']= employeedf_from_file.empSalary *2.4
employeedf_from_file
```

```python
#drop
employeedf_from_file.drop(['New Salary'],axis = 1, inplace = True)
employeedf_from_file
```

```python
dept = ['HR','OPS','DEV','','']
employeedf_from_file['DEPT']=dept
employeedf_from_file
```

```python
def incrementSalary(salary): 
    if salary <= 1500:
        newSal = salary * 1.1
    elif salary <= 10000:
         newSal = salary * 1.05
    elif salary <= 20000:
        newSal = salary * 1.025
    else:
         newSal = salary
    return int(newSal)
```

```python
#apply a function to a column:
employeedf_from_file['UpdatedYearlySalary']=employeedf_from_file['empSalary'].apply(incrementSalary)
employeedf_from_file
```

```python
#store data into csvfile
employeedf_from_file.to_csv('output.csv',index=None)
```
