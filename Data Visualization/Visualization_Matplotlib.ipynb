```python
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
avg_monthly_rain=[0,0,1,12,113,868,995,513,252,125,31,17]
month=list(range(1,13))
print(avg_monthly_rain)
print(month)
```

```python
plt.plot(month,avg_monthly_rain);
```

#### Documentation -> https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
plt.xlabel('Months')
plt.ylabel('Rainfall (in mm)')
plt.title('Monthly Rainfall Trend')
#plt.plot(month,avg_monthly_rain,linewidth=5, linestyle='dotted')
plt.plot(month,avg_monthly_rain,linewidth=5,linestyle='-.');
```

```python
max_rain=[15,10,19,32,213,968,1195,613,352,225,71,47]
```

```python
plt.xlabel('Months')
plt.ylabel('Rainfall (in mm)')
plt.title('Monthly Rainfall Trend')
plt.plot(month,max_rain,linewidth=3)
plt.plot(month,avg_monthly_rain,linewidth=5);
```

```python
plt.xlabel('Months')
plt.ylabel('Rainfall (in mm)')
plt.title('Monthly Rainfall Trend')
plt.plot(month,max_rain,linewidth=3,label='Maximum Rain')
plt.plot(month,avg_monthly_rain,linewidth=6,label='Average Rain')
plt.legend();
```

```python
plt.xlabel('Months')
plt.ylabel('Rainfall (in mm)')
plt.title('Monthly Rainfall Trend in Goa')
plt.plot(month,max_rain,linewidth=5, color='green',label='Maximum Rain')
plt.plot(month,avg_monthly_rain,linewidth=7,label='Average Rain')
plt.legend(loc='upper left', fontsize="large",shadow=True);
#loc=0 -->best location , 1-->upper right ,....
```

### Bar Chart

<h3>Simple bar chart showing revenues of tech companies</h3>

```python
company=['Infosys','TCS','Wipro','Hcl']
revenue=[190,236,189,97]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue);
```

```python
plt.bar(company,revenue, label="Revenue")
plt.ylabel("Revenue")
plt.xlabel("Company Name")
plt.title('Revenue Distribution for Top 4 Technology Companies')
plt.legend();
```

### Multiple Bars showing revenue and profit of Top tech companies

```python
profit=[40,22,15,7]
```

```python
plt.bar(company,revenue)
plt.bar(company,profit)
```

```python
plt.bar(xpos-0.15,revenue,width=0.3,label="Revenue")
plt.bar(xpos+0.15,profit,width=0.3,label="Profit")
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('Top Technology Companies')
plt.legend();
```

```python
plt.bar(company,revenue)
plt.bar(company,profit,align='edge') 
#edge-aligns the left edge of the bar (not the center, as it is by default) with the x position
plt.show()
```

```python
plt.barh(xpos-0.15,revenue, label="Revenue")
plt.barh(xpos+0.15,profit,label="Profit")
plt.yticks(xpos,company)
plt.xlabel("Revenue")
plt.title('Top Technology Companies')
plt.legend();
```

## Histogram

#### In histograms, x axis contains a variable and y axis will be a frequency of that variable

#### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
import numpy as np
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,rwidth=0.8) # by default number of bins is set to 10 
#rwidth - relative width as a fraction of the bin width
```

```python
min(blood_sugar)
```

```python
max(blood_sugar)
```

```python
np.linspace(min(blood_sugar),max(blood_sugar),11)
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.4)
```

```python
np.linspace(min(blood_sugar),max(blood_sugar),4)
```

<h3 style='color:black'>Histogram showing normal, pre-diabetic and diabetic patients distribution</h3>

<ul style="color:brown;font-weight: bold;">
    <li>80-100  : Normal</li>
    <li>101-125 : Pre-diabetic</li>
    <li>126-200  : Diabetic</li>
</ul>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
plt.hist(blood_sugar, bins=[80,100,125,200], rwidth=0.4,color=[0.2,0.5,0.3]);
```

#### Horizontal Orientation

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,200], rwidth=0.95,orientation='horizontal');
```

## Pie Chart 

#### Use Pie chart to analyze monthly home expenses

```python
exp_vals = [20000,10000,2000,4010,7500]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels);
```

```python
# percentage and explode8
plt.pie(exp_vals,labels=exp_labels, autopct='%.1f%%',radius=1.5,explode=[0,0,0,0,0.2]);
```

### Exporting the Plot as an image file
