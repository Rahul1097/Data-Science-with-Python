## Assumptions of Linear Regression

* Linear Relation : The relationship b/w Dependent & Independent is supposed to be linear. Dependent variable is a linear function of independent variable & error term
* Errors [Residuals] are Normally Distributed i.e. Regression Model is Robust
* Homoscedasticity : Equal variance in the errors across dataset
* Multicollinearity : Independence of the independent varaibles 

### 1. Linearity Assumption

```python
import matplotlib.pyplot as plt

X = train[['RM']] # is a dataframe 
y = train.MEDV # is a Series 
lm.fit(X, y)
```

```python
y_cap = lm.predict(X)
plt.figure(figsize = (10,5))
plt.scatter(train.RM, train.MEDV)
plt.plot(train.RM, y_cap,color = 'red')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.show();
```

### 2. Normality of Errors

**Residual Analysis  :**

* Residuals = y - y_cap 

```python
import scipy.stats as stats
```

```python
# Arranging and calculating the Residuals
residuals = pd.DataFrame({
    'actual values' : train.MEDV,
    'predicted values' : y_cap})

residuals['residuals'] = residuals['actual values'] - residuals['predicted values']
residuals.head()
```

```python
# Histogram for distribution
plt.figure(figsize=(8, 4))
plt.hist(residuals.residuals, bins = 150)
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Distribution of Error Terms')
plt.show()
```

### 3. Homoscedasticity 
* constant variance of residuals

### 4. Multicollinearity

* VIF - Variance Inflation Factor
* VIF(x) = 1 / (1 - R-Squared(X))
* R-Squared(X) is the R-Squared value coming from regression model for x i.e. taking X as dependent and all other variables as the independent variables

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
```

```python
# vif requires a data with intercept column
#the model will not be able to take it itself that's why intercept is added
x_data = train.drop(columns= 'MEDV')
x_data['Intercept'] = 1
x_data
```

```python
# creates an empty data frame
VIF = pd.DataFrame()
# creates a column named Independent_Variables that contains all the columns names (CRIM, ZN, INDUS,.....intercept).
VIF['Independent_Variables'] = x_data.columns
# Creates a new column VIF in which computes all the VIF values for all 12 columns
# (x_data.shape[1] returns the no of columns)
VIF['VIF'] = [vif(x_data.values, i) for i in range(x_data.shape[1])]

# changes the indexes from numbers to the name of the columns (instead of index 0 will have index CRIM, 
# instead of index 1 will have index ZN and so on), removes the column with the index "Intercept" 
# and transpose the vector(.T)
VIF = VIF.set_index('Independent_Variables').drop(index = 'Intercept').T
```
